## First we'll check Sales from Every Segments of Whole Data

```python
df=sample.groupby('Sub-Category')['Segment'].value_counts()
df.unstack('Segment').plot.bar(figsize=(15,10),stacked=True)
plt.legend(loc='best')
plt.ylabel('Counts')
plt.title("Sales From Every Segment")
plt.show();
```

### Inference: From above Graph we can say that "Home Office" segment has been less purchased Sub-Categories and in that "Tables","Supplies","Machines","Copiers","Bookcases" has the lowest Sales. "Consumer" has purchased more sub-categories as compared to other segments.

## Discounts given to various segments

```python
fig=plt.figure(figsize=(15,10))

df=sample.groupby('Sub-Category')[('Segment','Discount')].value_counts()

fig.add_subplot(3,1,1);   # 3 parameters : (row,column,location)
df.unstack('Segment').plot.bar(figsize=((15,4)),y='Consumer',rot=90)
plt.legend(loc='best')
plt.ylabel('Counts')
plt.title("Discounts to Consumer Segment Of United States of Whole Data")
plt.show();


fig.add_subplot(3,1,2);
df.unstack('Segment').plot.bar(figsize=((15,4)),y='Corporate',rot=90)
plt.legend(loc='best')
plt.ylabel('Counts')
plt.title("Discounts to Corporate Segment Of United States of Whole Data")
plt.show();

fig.add_subplot(3,1,3);
df.unstack('Segment').plot.bar(figsize=((15,4)),y='Home Office',rot=90)
plt.legend(loc='best')
plt.ylabel('Counts')
plt.title("Discounts to Home Office Segment Of United States of Whole Data")
plt.show();
```

```python
df=sample.groupby('Sub-Category')[('Segment','Discount')].value_counts()
df.unstack('Segment').plot.bar(figsize=(15,10),stacked=True)
plt.legend(loc='best')
plt.ylabel('Counts')
plt.title("Sales From Every Segment Of United States of Whole Data")
plt.show();
```

### Inference: Most of the discounts were given to Consumer Segment

## Relationship between Discount and Profit
